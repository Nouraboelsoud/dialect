```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import re
```

```python
df = pd.read_csv('dialect_data.csv')
df.head()
```

```python
def remove_user(text):
    return re.sub('(@\w+)*','' ,text)

def remove_eng(text):
    return re.sub('([a-z|A-Z]+\s*)*','' ,text)

def Extract_URL(text):
    m = re.search("(?P<url>https?://[^\s]+)", text)
    if m:
        return m.group("url")
def remove_emoji(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

def remove_num(text):
    text=re.sub(r'[0-9]+', '', text)
    num_pattern =re.compile("["
                            u"\u0660"
                            u"\u0661"
                            u"\u0662"
                            u"\u0663"
                            u"\u0664"
                            u"\u0665"
                            u"\u0666"
                            u"\u0667"
                            u"\u0668"
                            u"\u0669"
                            "]+", flags=re.UNICODE)
    return num_pattern.sub(r'', text)

def remove_special(text):
    special_pattern =re.compile("["
                            u"\u066A"
                            u"\u061F"
                            u"\u0021"
                            u"\u002E"
                            u"\u060C"
                            u"\u2025"
                            u"\u2026"
                            u"\u0023"
                            u"\u0025"
                            u"\u0028"
                            u"\u0029"
                            u"\u005F"
                            u"\u005E"
                            u"\u002D"                            
                            "]+", flags=re.UNICODE)
    return special_pattern.sub(r' ', text)

```

```python
df['clean_user']= df['DateValue'].apply(lambda x:remove_user(x))
df['clean_user']= df['clean_user'].apply(lambda x:remove_eng(x))
df['URL']= df['clean_user'].apply(lambda x:Extract_URL(x))
df['clean_user'] = df['clean_user'].apply(lambda x: x.replace(Extract_URL(x),'') if Extract_URL(x) else x)
df['clean_user']= df['clean_user'].apply(lambda x:remove_emoji(x))
df['clean_user']= df['clean_user'].apply(lambda x:remove_num(x))
df['clean_user']= df['clean_user'].apply(lambda x:remove_special(x))
df['clean_user']= df['clean_user'].apply(lambda x:x.replace('\n',' ').strip())

```
